```python
import pandas as pd 
```

```python
#import file
df = pd.read_csv('full_data.csv')
```

```python
#change nan to 0
df = df.fillna(0)
df
```

```python
#groupby location, get max no. of total cases, sort descending, choose top 11, convert series index to list
countries = df.groupby("location").total_cases.max().sort_values(ascending=False)[2:11].index.tolist()

#filter df for location in countries
df_filtered = df[df.location == "Italy"]
df_filtered = df_filtered[df_filtered.date > "2020-03-01"]
df_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import math
```
